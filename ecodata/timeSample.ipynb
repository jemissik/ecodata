```python
#MOD13A1.006__500m_16_days_NDVI_doy####%%%_aid0001.tif
# working code. Writing to code

import os
import io
import panel as pn
pn.extension('tabulator', 'terminal', 'ipywidgets', sizing_mode='stretch_width', loading_spinner='dots')
from tkinter import Tk, filedialog
import ecodata as eco
from ecodata.raster_utils import geotif2nc

# Create a FileSelector widget with an initial path
files = pn.widgets.FileSelector('~')
file_format_input = pn.widgets.TextInput(placeholder="Enter file format")
selected_directory = ""
file_list = []
file_output_path = ""

# Create a button to indicate that the user has finished selecting
finished_button = pn.widgets.Button(name="Finished Selecting")
callMethodButton = pn.widgets.Button(name="Call method")

# Create a text pane to display the selected directory
selected_directory_pane = pn.pane.Str()

# Create a text pane to display the selected output file
selected_output_file_pane = pn.pane.Str()
fileFormat = pn.pane.Str()

# Define a function to print the selected directory
def print_selected_directory(event):
    global selected_directory
    selected_directory = files.value
    print("Inside print_selected_directory method", flush=True)
    selected_directory_pane.object = f"Selected directory: {selected_directory}"
    file_list = os.listdir(selected_directory[0])

# Define a function to print the user input from the text box
def print_user_input(event):
    user_input = file_format_input.value
    print("Inside print_user_input method", flush=True)
    # selected_output_file_pane.object = f"User Input: {user_input}"

# Define a function to list all files in the selected directory
def list_files_in_directory(event):
    global file_output_path  # Use the global variable
    if selected_directory:
        print("inside selected directory condition", flush=True)
        try:
            content_to_append = "This text will be appended to the file."
            file_list = os.listdir(selected_directory[0])
            selected_output_file_pane.object = f"Files in the directory: {file_list}"
            if file_format_input.value:
                if file_output_path:
                    print("Inside the list_files_in_directory method")
                    geotif2nc(selected_directory[0], file_output_path, file_format_input.value)
        except FileNotFoundError:
            selected_output_file_pane.object = "Directory not found."
    else:
        selected_output_file_pane.object = "No directory selected."

# Link the button click events to the functions
finished_button.on_click(print_selected_directory)
finished_button.on_click(print_user_input)

fileOut = pn.pane.Str()

def select_files(*b):
    global file_output_path  # Use the global variable
    root = Tk()
    root.withdraw()
    root.call('wm', 'attributes', '.', '-topmost', True)
    files = filedialog.askopenfilename(multiple=False)
    file_output_path = files  # Update the global variable
    print("Inside selected_files method", flush=True)
    fileOut.object = f"Output File: {files}"
    fileFormat.object = f"File format: {file_format_input.value}"

button = pn.widgets.Button(name="Load file")
button.on_click(select_files)

# def toFile(event):
#     global file_output_path  # Use the global variable
#     if file_output_path:
#         file_object = open(file_output_path, 'w')
#         file_object.write('Checking if the file is being written.')
#         file_object.close()
#         print("The file should be closed.")

callMethodButton.on_click(list_files_in_directory)
# callMethodButton.on_click(toFile)

# Create a Panel dashboard with the FileSelector, Finished button, and the selected directory displayed
dashboard = pn.Column(files, finished_button, selected_directory_pane, button, fileOut, file_format_input, selected_output_file_pane, fileFormat, callMethodButton)

dashboard.servable()

#don't need toask for output file. Theu should pick a directory and name of the file. NetCDF. 
```
